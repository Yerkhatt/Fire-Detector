```python
import numpy as np
import pandas as pd
import tensorflow as tf
import os
from PIL import Image

HEIGHT, WIDTH, CHANNELS = 250, 250, 3
```

```python
def file_list(myDir, format='.jpg'):
    file_names = []
    
    print(myDir)
    for root, dirs, files in os.walk(myDir, topdown=False):
        for name in files:
            if name.endswith(format):
                file_names.append(name)
    return root, file_names
```

```python
test_X_root, test_X_list = file_list('../input/ds-fire-detection/test_imgs/test_imgs')
```

```python
# load model
ld_model = tf.keras.models.load_model(filepath = '../input/trained-model/FireDectector.h5')
```

```python
def prediction(root, FileList):
    pred = []
    for file in FileList:
        img_file = Image.open(os.path.join(root, file))
        
        img_file = img_file.resize((HEIGHT, WIDTH))
        
        value = np.array(img_file.getdata())/255
        value = value.reshape(1, HEIGHT, WIDTH, CHANNELS)
        #print(value.shape)
        p = ld_model.predict(value)
        
        pred.append([str(file), round(p[0][0])])
    
    
    return np.array(pred)
```

```python
pred_test = prediction(test_X_root, test_X_list)
```
